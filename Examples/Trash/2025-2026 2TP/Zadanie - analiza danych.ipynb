### Oczyszczanie danych

W pliku znajdują się kolumny:

- Id – unikalny identyfikator (niepotrzebny w analizie)
- imie
- nazwisko (również dwuczłonowe)

- roczne_zarobki – kwota + waluta (np. „12345.67 PLN”, „9876,50 EUR”, „45000 USD”). Mogą znajdować się również kwoty bez walut

Dane zawierają błędy, które należy usunąć lub poprawić.

**Wymagane operacje:**  
- Usunięcie zbędnych danych
    - Usuń całą kolumnę Id.
    - Usuń rekordy, w których pole roczne_zarobki jest puste.

**Czyszczenie pól tekstowych:**
- Usuń niepotrzebne spacje na początku i końcu pól imie i nazwisko.
- Dostosuj wielkość liter: imię i nazwisko powinny zaczynać się wielką literą, pozostałe litery małe. Uwaga na nazwiska dwuczłonowe

**Walidacja numeru PESEL**
- Sprawdź, czy PESEL ma dokładnie 11 znaków. Jeśli PESEL jest niepoprawny → ustaw wartość NULL. Nie zwracaj uwagi na poprawność cyfr w numerze PESEL, obecnie są to wartości losowe, nie zawierają sensownych danych

````python
# definicje dodatkowych funkcji

def popraw(arg:str):
    arg = f'{arg}'
    arg = arg.replace(' ','')
    arg = arg.capitalize()
    return arg

def popraw_n(arg:str):
    arg = f'{arg}'
    frag = arg.split('-')
    if len(frag) > 1:
        return f'{popraw(frag[0])}-{popraw(frag[1])}'
    else:
        return popraw(arg)

def PESEL(arg):
    arg = f'{arg}'
    if len(arg) != 11:
        return None
    else:
        return arg

# def popraw(arg:str):
#     arg = f'{arg}'
#     return arg.replace(' ','').title()
````

````python
# dostosowanie wielkości liter w kolumnie "imie" do formatu "Imię" (pierwsza litera wielka, reszta małe)
df = kopia2
df['imie'] = df['imie'].apply(popraw)
df['nazwisko'] = df['nazwisko'].apply(popraw_n)
df
````

````python
# walidacja nr PESEL

df['PESEL'] = df['PESEL'].apply(PESEL)
df
````

### Przetwarzanie danych finansowych  

Pole roczne_zarobki zawiera kwotę oraz walutę. Występują różne formaty liczb (kropka lub przecinek jako separator dziesiętny).

**Dodanie kolumny kod_kraju**  
- Na podstawie waluty dodaj nową kolumnę:  
   - Waluta → Kraj  
   - PLN	→ POL  
   - USD	→ USA  
   - EUR	→ GER  

**Konwersja zarobków do jednej waluty (PLN)** 
- Zamień przecinek na kropkę, jeśli występuje.
- Wyodrębnij walutę (PLN, USD, EUR).
- Jeśli waluta nie występuje → przyjmij, że jest to PLN.
- Przelicz wszystkie kwoty na PLN według ustalonych kursów (stan na 5.03.2026 r.):
    - 1 USD = przyjęty kurs 3.677 PLN
    - 1 EUR = przyjęty kurs 4.268 PLN
    - PLN pozostaje bez zmian

Wynik zapisz w nowej kolumnie, np. zarobki_PLN.

````python
# dodatkowe funkcje
def nowa_k(arg):
    arg = f'{arg}'
    last = arg[-3:]

    if last == 'PLN':
        return 'POL'
    elif last == 'USD':
        return 'USA'
    elif last == 'EUR':
        return 'GER'
    else:
        return 'POL'
    
def val(arg):
    arg = f'{arg}'
    arg = arg.replace(',','.')
    arg = arg.replace(' ','')
    waluta = arg[-3:]
    mn = 1
    if waluta == 'USD':
        mn = 3.677
    elif waluta == 'EUR':
        mn = 4.268
   

    try:
        return mn * float(arg)
    except:
        try:
            return mn * float(arg[:-3])
        except:
            return float(0)
    
````

````python
# dodanie kolumny kod_kraju na podstawie waluty (np. "PLN" -> "POL", "USD" -> "USA", "EUR" -> "GER")
df['kod_kraju'] = df['roczne_zarobki'].apply(nowa_k)
df
````

````python
# przetworzenie kolumny roczne_zarobki, usunięcie symbolu waluty i konwersja na typ numeryczny

df['roczne_zarobki'] = df['roczne_zarobki'].apply(val)
````
